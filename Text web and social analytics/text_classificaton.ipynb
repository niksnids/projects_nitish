```python
#5.
vectorizer = TfidfVectorizer(max_df=0.7, min_df=0.1)
X = vectorizer.fit_transform(df.preprocessed)
```

```python
data_stem2 = pd.DataFrame.sparse.from_spmatrix(X)
```

```python
df = df.reset_index()
```

```python
data_stem2["target"] = df["target"]
data_stem2["target_names"] = df["target_names"]
```

#Part B

```python
#6.
def text_train(df):
  return train_test_split(df.iloc[:, :-2], df.target,test_size = 0.20, random_state = 12)
```

```python
docs_train_s, docs_test_s, y_train_s, y_test_s = text_train(data_stem2)
docs_train_w2v, docs_test_w2v, y_train_w2v, y_test_w2v = text_train(data_w2v)
docs_train_Bert, docs_test_Bert, y_train_Bert, y_test_Bert = text_train(data_BERT)
```

It splits the data so one can apply supervised learning (having a training model and then test the performance of the trained model on a test dataset).

```python
clf = MultinomialNB()
clf.fit(docs_train_s, y_train_s)
```

It is a Multinomial Naive Bayes classifier which is trained on the stemmed dataset.

```python
y_pred_s = clf.predict(docs_test_s)
```

It predicts on the test data with the above model.

```python
print(clf.score(docs_train_s, y_train_s))
print(accuracy_score(y_pred_s, y_test_s))
print(classification_report(y_pred_s, y_test_s))
```

Score: The mean accuracy of the training model.

Accuracy Score: Confusion Matrix based approach where the true positives are measured against the actual positives.

Classification Report: Summarizes everything.

It depends on the use case. Can be a good model but also can't be sufficient depending on how you define your model to be.

```python
clf2 = RandomForestClassifier(random_state = 42)
clf2.fit(docs_train_s, y_train_s)
```

```python
y_pred_s_rf = clf2.predict(docs_test_s)
```

Looks better.

```python
clf3 = RandomForestClassifier(random_state = 42)
clf3.fit(docs_train_w2v, y_train_w2v)
y_pred_w2v_rf = clf3.predict(docs_test_w2v)

clf4 = RandomForestClassifier(random_state = 42)
clf4.fit(docs_train_Bert, y_train_Bert)
y_pred_Bert_rf = clf4.predict(docs_test_Bert)
```

```python
print("W2V")
print(clf3.score(docs_train_w2v, y_train_w2v))
print(accuracy_score(y_pred_w2v_rf, y_test_w2v))
print(classification_report(y_pred_w2v_rf, y_test_w2v))

print("Bert")
print(clf4.score(docs_train_Bert, y_train_Bert))
print(accuracy_score(y_pred_Bert_rf, y_test_Bert))
print(classification_report(y_pred_Bert_rf, y_test_Bert))
```

Severe overfitting of the w2c model.

#Part D

```python
param_grid = {'min_samples_leaf': [5,10], 'n_estimators': [3,5]}
rf = RandomForestClassifier(random_state = 42)
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 10)
grid_search.fit(docs_train_s, y_train_s)
```

```python
grid_search.best_params_
```

```python
bestmodel = grid_search.best_estimator_.fit(docs_train_s, y_train_s)
y_pred_s_grid = bestmodel.predict(docs_test_s)
```

```python
print(bestmodel.score(docs_train_s, y_train_s))
print(accuracy_score(y_pred_s_grid, y_test_s))
print(classification_report(y_pred_s_grid, y_test_s))
```
